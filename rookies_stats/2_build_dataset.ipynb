```python
import pandas as pd
```

```python
leagues = ['Serie-B', 'La-Liga', 'Ligue-1', 'Premier-League', 'Bundesliga', 'Eredivisie', 'Primera-Division']
seasons = ['2023-2024']
```

```python
start = 0
for i in range( len(leagues) ):
    for j in range( len(seasons) ):
        exception = 0
        try:
            rcsv = pd.read_csv('fbref_data/' + leagues[i] + '_' + seasons[j] + '_outfield.csv')   
        except:
            exception = 1
        if(exception):
            continue
            
        outfield_players = pd.DataFrame(rcsv)
        
        
        
        
        rcsv = pd.read_csv('fbref_data/' + leagues[i] + '_' + seasons[j] + '_team.csv')   
        team = pd.DataFrame(rcsv)
        
        rcsv = pd.read_csv('fbref_data/' + leagues[i] + '_' + seasons[j] + '_teamvs.csv')   
        vs_team = pd.DataFrame(rcsv)
        
        for k in range(1, team.shape[1]):
            team = team.rename(columns = {team.columns[k] : 'team_' + team.columns[k]})
    
        for k in range(1, vs_team.shape[1]):
            vs_team = vs_team.rename(columns = {vs_team.columns[k] : 'vs_team_' + vs_team.columns[k]})

        vs_team.pop('team')

        team_data = pd.concat([team, vs_team], axis = 1)

        team_data['team_idx'] = team_data['team']

        team_data = team_data.set_index('team_idx')
        
        
        outfield_players = pd.concat( [outfield_players.reset_index(drop = True), team_data.loc[outfield_players['team']].reset_index(drop = True) ], axis = 1)
        

        outfield_players['league'] = outfield_players['player']
        outfield_players['season'] = outfield_players['player']

        outfield_players.loc[:, 'league'] = leagues[i]
        outfield_players.loc[:, 'season'] = seasons[j]
            
        
        if(start == 0):
            players = outfield_players
            start = 1
        else:
            cols = players.columns.intersection(outfield_players.columns)
            players = pd.concat(   [ players[cols].reset_index(drop = True), outfield_players[cols].reset_index(drop = True) ], axis = 0, ignore_index = True)

                
```

```python
import unicodedata

def normalize_name(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    return only_ascii.decode('utf-8')
```

```python
players['surname'] = players['player']
players['initial'] = players['player']

for i in range(players.shape[0]):
    players.loc[i, 'surname'] = players.loc[i, 'surname'].split(' ')[-1]
    players.loc[i, 'surname'] = normalize_name(players.loc[i, 'surname'].replace('\'', ''))
    
    
    players.loc[i, 'initial'] = players.loc[i, 'player'][0]
```
